```python
import torch

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
# Images
imgs = ['/content/imagenes/tattoo.jpg']  # batch of images
```

```python
from PIL import Image
img = Image.open(r'/content/imagenes/tattoo.jpg')
img
```

```python
# Inference
results = model(imgs)
# Results
results.print()
```

```python
# Results
results.print()
```

```python
results.save()  # or .show()
```
